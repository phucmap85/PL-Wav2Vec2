```python
import librosa
import os
from tqdm import tqdm

def check_sampling_rates(root_path):
    sampling_rates = {}
    audio_files = []

    for root, dirs, files in os.walk(root_path):
        for file in files:
            if file.endswith(('.wav', '.mp3', '.flac')):
                audio_files.append(os.path.join(root, file))

    print(f"Tìm thấy {len(audio_files)} tệp âm thanh. Đang kiểm tra...")

    for path in tqdm(audio_files):
        try:
            sr = librosa.get_samplerate(path)
            sampling_rates[sr] = sampling_rates.get(sr, 0) + 1
        except Exception as e:
            print(f"Lỗi khi đọc {path}: {e}")

    return sampling_rates

root_dir = '/content/audio_data'
results = check_sampling_rates(root_dir)

print("\nKết quả thống kê tần số lấy mẫu:")
for sr, count in results.items():
    print(f"- {sr} Hz: {count} tệp")
```

```python
import librosa
import soundfile as sf
import os
from tqdm import tqdm

def trim_silence_from_dir(root_path, top_db=30):
    audio_files = []
    for root, dirs, files in os.walk(root_path):
        for file in files:
            if file.endswith(('.wav', '.mp3', '.flac')):
                audio_files.append(os.path.join(root, file))

    print(f"Bắt đầu xử lý cắt khoảng lặng cho {len(audio_files)} tệp...")

    for path in tqdm(audio_files):
        try:
            y, sr = librosa.load(path, sr=None)

            y_trimmed, index = librosa.effects.trim(y, top_db=top_db)

            sf.write(path, y_trimmed, sr)
        except Exception as e:
            print(f"Lỗi xử lý {path}: {e}")

root_dir = '/content/audio_data'
trim_silence_from_dir(root_dir)
print("\nĐã hoàn thành cắt khoảng lặng cho toàn bộ dữ liệu.")
```
